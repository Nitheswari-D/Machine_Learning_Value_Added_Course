## LINE GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,10])
y=np.array([0,100])
plt.plot(x,y)
plt.show()
```

## Plot function

It draws a line from one point to another point.
It has two parameters first paramater is x axis points and second parameter is y axis points

```python
x=np.array([0,10,20,30])
y=np.array([0,100,200,300])
plt.plot(x,y,'o')
```

```python
x=np.array([2,3,8,10])
y=np.array([4,9,3,12])
plt.plot(x,y)
#by default x is not given so it takes as 1,2,3,4....
y=np.array([2,3,8,10])
plt.plot(y)
```

```python
#marker is used to highlight the intersection (x,y) points 
y=np.array([2,3,8,10])
plt.plot(y,marker='o') #dot
```

```python
import numpy as np
import matplotlib.pyplot as plt
y=np.array([2,3,8,10])
plt.plot(y,marker='*') # star
```

```python
y=np.array([2,3,8,10])
plt.plot(y,marker='+') #plus
```

## SCATTER PLOT

```python
import numpy as np
import matplotlib.pyplot as plt
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
plt.scatter(x1,y1)
plt.title('India')
plt.show()
```

```python
#default colors
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
plt.scatter(x1,y1)
x2=np.array([1,2,3,4,5])
y2=np.array([10,7,16,19,2])
plt.scatter(x2,y2)
plt.title('Ind vs Aus')
plt.show()
```

```python
#colors mentioned
plt.scatter(x1,y1,c='r') 
plt.scatter(x2,y2,c='g')
```

```python
#specific color for each (x,y) i.e (1,14) => red
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
c1=np.array(['red','green','blue','orange','magenta'])
plt.scatter(x1,y1,c=c1)
x2=np.array([1,2,3,4,5])
y2=np.array([10,7,16,19,2])
c2=np.array(['cyan','black','grey','brown','pink'])
plt.scatter(x2,y2,c=c2)
plt.title('Ind vs Aus')
plt.show()
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
c1=np.array(['red','green','blue','orange','magenta'])
plt.scatter(x1,y1,c=c1,cmap='viridis') #virtual display
plt.colorbar()
plt.show()
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
c1=np.array(['red','green','blue','orange','magenta'])
plt.scatter(x1,y1,c=c1,cmap='Accent')
plt.colorbar()
plt.show()
```

```python
#creating two arrays and color array using randomized integer
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
plt.scatter(x, y, c=colors)
plt.show()
```

```python
#size used to display the (x,y) points in different sizes
x = np.random.randint(10, size=(10))
y = np.random.randint(10, size=(10))
colors = np.random.randint(10, size=(10))
size=np.array([10,15,20,50,100,30,40,25,55,60])
plt.scatter(x, y, c=colors,s=size)
```

```python
#opacity for (x,y) points
plt.scatter(x,y,c=colors,s=size,alpha=0.5)
```

## BAR CHART

```python
a=np.array([1,2,3,4,5])
b=np.array([15,9,11,19,8])
plt.bar(a,b)
plt.show()
```

```python
# for horizontal representation of bar graph using barh()
a=np.array([1,2,3,4,5])
b=np.array([15,9,11,19,8])
plt.barh(a,b)
plt.show()
```

```python
#To change the color of the bar chart plt.bar(a,b,"color")
a=np.array([1,2,3,4,5])
b=np.array([15,9,11,19,8])
plt.bar(a,b,color='pink')
plt.show()
```

```python
# To change the size or width of the bars in graph
# width is used in vertical representation of the graph
a=np.array([1,2,3,4,5])
b=np.array([15,9,11,19,8])
plt.bar(a,b,color='pink',width=0.2)
plt.show()
```

```python
# To change the height of the bars in the graph 
#height is used in horizontal bar representation i.e barh()
a=np.array([1,2,3,4,5])
b=np.array([15,9,11,19,8])
plt.barh(a,b,color='pink',height=0.3)
plt.show()
```

## PIE CHART

```python
a=np.array([40,25,15,20])
plt.pie(a)
plt.show()
```

```python
#Giving labels to each part of pie chart
a=np.array([40,25,15,20])
l=np.array(["Ind","Aus","Eng","Pak"])
plt.pie(a,labels=l)
plt.show()
```

```python
#rotating the pie chart by specific angle
a=np.array([40,25,15,20])
l=np.array(["Ind","Aus","Eng","Pak"])
plt.pie(a,labels=l,startangle=180)
plt.show()
```

```python
a=np.array([40,25,15,20])
l=np.array(["Ind","Aus","Eng","Pak"])
ex=[0.05,0.05,0.05,0.05]
plt.pie(a,labels=l,startangle=180,explode=ex)
plt.show()
```

```python
a=np.array([40,25,15,20])
l=np.array(["Ind","Aus","Eng","Pak"])
ex=[0.05,0.05,0.05,0.05]
c=["black","r","#00ff00","pink"]
plt.pie(a,labels=l,startangle=180,explode=ex,colors=c)
plt.show()
```

```python
a=np.array([40,25,15,20])
l=["Ind","Aus","Eng","Pak"]
c=["black","r","#00ff00","pink"]
plt.pie(a,labels=l,startangle=180,colors=c)
plt.legend(title="Scores")
plt.show()
```
